```python
import pandas as pd
import skmob
from skmob.preprocessing import compression
import warnings
warnings.filterwarnings('ignore')
```

```python
sample_filtered = pd.read_csv('./train_filtered_sample.csv')
```

```python
# Pega uma trajetória como exemplo
trip_id = trip_id_unique = sample_filtered['trip_id'].unique()[1]
one_trip = sample_filtered[sample_filtered['trip_id'] == trip_id]
# Transforma trajetória em um objeto TrajDataFrame
one_tdf = skmob.TrajDataFrame(one_trip, latitude='lat', longitude='lon', user_id='taxi_id', datetime='timestamp', crs={'init': 'WGS84'})
# Comprime trajetória
one_tdf_compressed = compression.compress(one_tdf)
```

```python
def plot_trajectory_diff(tdf: skmob.TrajDataFrame, tdf2: skmob.TrajDataFrame, zoom = 15):
  diff_points = tdf[~tdf.apply(tuple,1).isin(tdf2.apply(tuple,1))]
  map_f = diff_points.plot_stops(hex_color='green', zoom = zoom)
  map_f = tdf.plot_trajectory(map_f=map_f, dashArray='5,5', start_end_markers=False, hex_color='red')
  map_f = tdf2.plot_trajectory(map_f=map_f, start_end_markers=False, hex_color='blue')
  return map_f
```

```python
plot_trajectory_diff(one_tdf, one_tdf_compressed)
```
